```python
#For example, in the ClinVar database, 1724 genetic variants are associated with "Hereditaary cancer predisposing syndrome".
[col for col in df.columns if 'sig' in col.lower()]
df['CLNDN_clean'] = df['CLNDN'].astype(str).str.split(':').str[-1]
df['CLNDN_clean'].value_counts()
```

```python
from scipy import stats
# any of the 3 t Tests discussed last week
# Filter two groups
df['CLNDN_clean'] = df['CLNDN'].astype(str).str.split(':').str[-1]
df['CLNDN_clean'].value_counts()
group1 = df[df['CLNDN_clean'].str.contains('Hereditary_cancer-predisposing_syndrome', na=False)]['AF_EXAC']
group2 = df[df['CLNDN_clean'].str.contains('not_specified', na=False)]['AF_EXAC']

# Drop NaN values
group1 = group1.dropna()
group2 = group2.dropna()

# Perform independent t-test
# NULL HYPOTHESIS: THERE IS NO DIFF IN THE MEAN AF BETWEEN VARIANTS ASSOCITED WITH HCPS & NSV
t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

# (5 points) Visualization
Create a figure or figures that visualize the information you wish to present.


```python
 import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.boxplot(
    [group1, group2],
    tick_labels=['Hereditary Cancer', 'Not Specified'],
    patch_artist=True
)
plt.title('Allele Frequency Comparison: Hereditary Cancer vs. Not Specified')
plt.ylabel('Allele Frequency (AF_EXAC)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import seaborn as sns


# Prepare a combined DataFrame for seaborn

data = pd.DataFrame({
    'Allele_Frequency': pd.concat([group1, group2]),
    'Group': ['Hereditary Cancer'] * len(group1) + ['Not Specified'] * len(group2)
})

plt.figure(figsize=(8,6))
sns.violinplot(x='Group', y='Allele_Frequency', hue='Group', data=data, inner='box', palette='Set2', legend=False)
plt.title('Allele Frequency Distribution by Group')
plt.ylabel('Allele Frequency (AF_EXAC)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(-0.05, .1)
plt.show()
```
